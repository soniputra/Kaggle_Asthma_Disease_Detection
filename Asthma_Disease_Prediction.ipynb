### About Dataset
This dataset contains extensive health information for 2,392 patients diagnosed with Asthma Disease, uniquely identified with IDs ranging from 5034 to 7425. It includes demographic details, lifestyle factors, environmental and allergy factors, medical history, clinical measurements, symptoms, and a diagnosis indicator. This dataset is valuable for researchers and data scientists aiming to explore factors associated with Asthma, develop predictive models, and conduct statistical analyses.

<h3>Patient Information</h3>
<b>Patient ID</b>
<ul><li><b>PatientID</b>: A unique identifier assigned to each patient (5034 to 7425).</ul></li>
<b>Demographic Details</b>
<ul>
    <li>Age: The age of the patients ranges from 5 to 80 years.</li>
    <li>Gender: Gender of the patients, where 0 represents Male and 1 represents Female.
    <li>Ethnicity: The ethnicity of the patients, coded as follows:
    <li>0: Caucasian
    <li>1: African American
    <li>2: Asian
    <li>3: Other
    <li>EducationLevel: The education level of the patients, coded as follows:
    <li>0: None
    <li>1: High School
    <li>2: Bachelor's
    <li>3: Higher
</ul>
<h3> Lifestyle Factors </h3>
<ul>
    <li>BMI: Body Mass Index of the patients, ranging from 15 to 40.
    <li>Smoking: Smoking status, where 0 indicates No and 1 indicates Yes.
    <li>PhysicalActivity: Weekly physical activity in hours, ranging from 0 to 10.
    <li>DietQuality: Diet quality score, ranging from 0 to 10.
    <li>SleepQuality: Sleep quality score, ranging from 4 to 10.
</ul>
<h3> Environmental and Allergy Factors </h3>
<ul>
    <li>PollutionExposure: Exposure to pollution, score from 0 to 10.
    <li>PollenExposure: Exposure to pollen, score from 0 to 10.
    <li>DustExposure: Exposure to dust, score from 0 to 10.
    <li>PetAllergy: Pet allergy status, where 0 indicates No and 1 indicates Yes.
</ul>
<h3> Medical History  </h3>
<ul>
    <li>FamilyHistoryAsthma: Family history of asthma, where 0 indicates No and 1 indicates Yes.
    <li>HistoryOfAllergies: History of allergies, where 0 indicates No and 1 indicates Yes.
    <li>Eczema: Presence of eczema, where 0 indicates No and 1 indicates Yes.
    <li>HayFever: Presence of hay fever, where 0 indicates No and 1 indicates Yes.
    <li>GastroesophagealReflux: Presence of gastroesophageal reflux, where 0 indicates No and 1 indicates Yes.
</ul>
<h3> Clinical Measurements </h3>
<ul>
    <li>LungFunctionFEV1: Forced Expiratory Volume in 1 second (FEV1), ranging from 1.0 to 4.0 liters.
    <li>LungFunctionFVC: Forced Vital Capacity (FVC), ranging from 1.5 to 6.0 liters.
</ul>
<h3> Symptoms </h3>
<ul>
    <li>Wheezing: Presence of wheezing, where 0 indicates No and 1 indicates Yes.
    <li>ShortnessOfBreath: Presence of shortness of breath, where 0 indicates No and 1 indicates Yes.
    <li>ChestTightness: Presence of chest tightness, where 0 indicates No and 1 indicates Yes.
    <li>Coughing: Presence of coughing, where 0 indicates No and 1 indicates Yes.
    <li>NighttimeSymptoms: Presence of nighttime symptoms, where 0 indicates No and 1 indicates Yes.
    <li>ExerciseInduced: Presence of symptoms induced by exercise, where 0 indicates No and 1 indicates Yes.
</ul>
<h3> Diagnosis Information </h3>
<ul>
    <li>Diagnosis: Diagnosis status for Asthma, where 0 indicates No and 1 indicates Yes.
</ul>
<h3> Confidential Information </h3>
<ul>
    <li>DoctorInCharge: This column contains confidential information about the doctor in charge, with "Dr_Confid" as the value for all patients.
</ul>

```python
# Library For Statistical
import pandas as pd
import numpy as np

# Library For Visualization
import matplotlib as mpl
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Library For Model Predict
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Library For Accuracy Result
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn.metrics import accuracy_score,f1_score, precision_score, recall_score, mean_squared_error

# Library for saving into the model
import pickle

import warnings
warnings.filterwarnings("ignore")
```

## Data Loading

### Import and Load The Data

```python
df = pd.read_csv("dataset/asthma_disease_data.csv")
df
```

### Data Copy

```python
df_copy = df.copy()
df_copy
```

## Data Manipulation

### Change The Data Type and Convert to only one decimal

```python
df_copy["BMI"] = df_copy["BMI"].astype("int64")
df_copy["PhysicalActivity"] = df_copy["PhysicalActivity"].astype("int64")
df_copy["DietQuality"] = df_copy["DietQuality"].astype("int64")
df_copy["SleepQuality"] = df_copy["SleepQuality"].astype("int64")
df_copy["PollutionExposure"] = df_copy["PollutionExposure"].astype("int64")
df_copy["PollenExposure"] = df_copy["PollenExposure"].astype("int64")
df_copy["DustExposure"] = df_copy["DustExposure"].astype("int64")

### Untuk Round Section
df_copy["LungFunctionFEV1"] = df_copy["LungFunctionFEV1"].round(1)
df_copy["LungFunctionFVC"] = df_copy["LungFunctionFVC"].round(1)
```

### Checking the result

```python
df_copy
```

```python
df_copy.info()
```

### Remove Unused Columns 

```python
df_copy = df_copy.drop(columns=["PatientID"])
df_copy = df_copy.drop(columns=["DoctorInCharge"])
```

### Check The Result

```python
df_copy
```

## Data Balancing

### Divide them into each category with a maximum number of data per category of 124.

```python
## Trying Balancing Technique

df_possdie = df_copy[df_copy["Diagnosis"] == 1].reset_index(drop=True)
df_possnondie = df_copy[df_copy["Diagnosis"] == 0].sample(124).reset_index(drop=True)
```

### Check the result

#### Possible to Die (Based on Diagnosis)

```python
df_possdie
```

#### Possible Not to Die (Based on Diagnosis)

```python
df_possnondie
```

### Merge the Data

```python
df_merged = pd.concat([
    df_possdie,
    df_possnondie
]).reset_index(drop=True)

df_merged
```

## Visualization

## Analyst Section / Prediction Model Development

```python
df_merged.info()
```

### Drop The Target Column

```python
X = df_merged.drop("Diagnosis", axis=1)
y = df_merged["Diagnosis"]
print(X.shape)
print(y.shape)
```

## Set Train and Test Data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=26)
```

### Set Model

```python
model1 = KNeighborsClassifier(n_neighbors=26)
```

```python
model1.fit(X_train, y_train)

y_pred11 = model1.predict(X_test)
# evaluasi model
confusion_matrix(y_test,y_pred11)

print(classification_report(y_test,y_pred11))
```

```python
print("=============================================================")
print('Model Conclusion KNN Classifier')
print("=============================================================")
print('Akurasi :',round(accuracy_score(y_test, y_pred11),4))
print('Presisi :',round(precision_score(y_test,y_pred11),4))
print('Recall :',round(recall_score(y_test,y_pred11),4))
print('F1 Score :',round(f1_score(y_test,y_pred11),4))
print("=============================================================")
```

```python
cm11 = confusion_matrix(y_test, y_pred11)
sns.heatmap(cm11, annot=True, fmt="d", linewidths=.5, cmap = 'Reds',)
```
